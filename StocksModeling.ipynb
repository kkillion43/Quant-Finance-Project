```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2,1)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[3], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[4], seasonal_pdq[4]))
print('SARIMAX: {} x {}'.format(pdq[5], seasonal_pdq[5]))
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages
bestResult = 999999

bestResults = []

arimaFrame = arimaMonthHigh.set_index('Dates')

for param in pdq:
    print("Going through Parameters: ", param)
    for param_seasonal in seasonal_pdq:
        for stock in arimaFrame:
            
            try:
                mod = sm.tsa.statespace.SARIMAX(arimaFrame[stock],
                                                order=param,
                                                seasonal_order=param_seasonal,
                                                enforce_stationarity=False,
                                                enforce_invertibility=False)

                results = mod.fit()
                
#                 print('{} : ARIMA{}x{}12 - AIC:{}'.format(stock, param, param_seasonal, results.aic))
                
                if results.aic < bestResult:
                    print(stock, results.aic, mod.order, param)
                    bestResults.append([stock, results.aic, results.bic, mod.order, mod.seasonal_order])
                    

            except Exception as e:
                print(str(e))
                continue

bestdf = pd.DataFrame(bestResults, columns=['Ticker', 'AIC', 'BIC', 'Parameters', 'Sesonal'])

print(bestdf.head(), '\n', bestdf.groupby('Ticker').describe(), '\n')

print('Lowest Results:\n')
bestdf.groupby('Ticker').min()
```

"The Regression results tells us alot, including Beta, Alpha and R^2 , beta is the market coefficient or SPY's coef , alpha is the constant coef or const coef since it's the distance between X axis and the line" (Yahya). 

```python
import statsmodels.api as sm

X = sm.add_constant(dfm[(dfm.Ticker == 'SPY') & (dfm.Returns < 100)]['Returns'] )

# y is the values of the monthly returns of the stock = the first column
y = dfm[(dfm.Ticker == 'C') & (dfm.Returns < 100)]['Returns']

# FIT the Model
model = sm.OLS(y,X).fit()

# Print Summary of the OLS Regression Results
print(model.summary())
```

```python
beta = model.params["Returns"]
alpha = model.params["const"]
r2 = model.rsquared
print("BETA: ", beta, "\nALPHA: ", alpha, "\nR-Squared: ", r2)
```
